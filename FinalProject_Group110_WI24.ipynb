# Abstract

**Introduction:**

The COVID-19 pandemic has brought forth unprecedented challenges, affecting target and at risk populations across the globe. It is already well known that pregnant women are significantly susceptible to increased mortality and morbidity due to infectious diseases (C. M Lam et al., 2004). Recent research has shown that this vulnerability is heightened by significant physiological changes during pregnancy throughout their immune system, thus increasing their risk of infection of diseases, including COVID-19 (Wastnedge et al., 2021). In addition, birth related fears have been connected to increased psychological distress with negative outcomes for both birth and infant development, with birth related fears being present amongst ~25% of a pregnant population that ~10% of also reported that the fear impaired daily activities and the ability to the cope with labor (Molgora et al., 2018; Ding et al., 2014; Vismara et al., 2020). Despite this research on fear and pregnancy, very little is known about how health-related fears, such as fear of COVID-19, affect a pregnant woman’s maternal mental health and childbirth outcomes. This lack of research is incredibly important because it limits a health system’s ability to combat the challenges faced by pregnant individuals, who are at risk when pandemics arise in the future.

This study, which leverages data from the Pregnancy during the COVID-19 Pandemic project, (PdP), had the goal of further dissecting the interplay between mental health disturbances and specific pandemic-related events and their subsequent effect if any on pregnancy outcomes. Using a variety of statistical tools, including pairplot visualizations, correction analyses, and logistic regression, our research hopes to shine light onto the relationships between perceived threat levels of COVID-19, anxiety, depression, and the likelihood of NICU admission. This study is focused on highlighting the significant, long lasting impacts of COVID-19, including investigating critical milestones such as the first national COVID-19 vaccine rollout, and the mental health of COVID-19 era pregnancies. Our exploratory data analysis spins over the pre and post-vaccine time periods, offering an in depth look at the shifting landscape of pregnant women’s mental health following the first vaccine rollout.

Our findings showcased consistent associations between perceived pandemic threat levels and markers of mental health distress across both of our observed periods. We found a significant positive correlation between perceived threat level and negative mental health indicators, highlighting the effectiveness of pandemic-related stressors on maternal mental well being. Despite this, our logistic regression found that perceived threats of COVID-19  are not significant predictors of NICU admissions. 

Though our overall goal of the study was to highlight the importance of mental health care for pregnant women during high stress inducing events such as COVID-19, our findings reinforces the need for more holistic health care strategies that include support for pregnant woman, and specifically address the risk of physiological changes and diseases during pregnancy and alleviate the perceived fears related to their pregnancy and birth. In conclusion, our findings show the importance of addressing mental health concerns among pregnant individuals, particularly during times of stress such as the COVID-19 pandemic. While the introduction of vaccines may have lessened some psychological distress, perceived threat levels and mental health indicators exhibited that there was still a need for maternal well-being.



We created a function named **standardize_income**, which takes in an input named income and standaredizes the income into numbers and removes symbols like "+". It then converts the income from a string to an integer so we can analyze it later on.

```python
#this function converts the income into integer values, taking the maximum of those values with ranges
def standardize_income(income):
    if pd.isna(income):
        return np.nan

    if 'Less than' in income:
        return 20000

    # Handle open-ended values like '$200,000+'
    elif '+' in income:
        return 200000

    # Check if income is a range
    else:
        parts = income.replace('$', '').replace(',', '').replace(' ', '').split('-')
        #convert to int
        int_parts = [int(x) for x in parts]
        # Calculate maximum for ranges
        return int_parts[1]

# Assuming 'df' is your dataframe
df['income'] = df['income'].apply(standardize_income)

df.income.value_counts()
```

```python
# Define a function to perform linear regression and print results
def perform_linear_regression(data):
    X = data[['depression', 'anxiety']]
    y = data['birth_weight_g']

    model = LinearRegression()
    model.fit(X, y)

    # Extracting model coefficients and intercept
    coefficients = model.coef_
    intercept = model.intercept_

    print(f'Regression coefficients: {coefficients}')
    print(f'Intercept: {intercept}')

# Applying the function to the prevaccine and postvaccine subsets
print("Pre-vaccine period:")
perform_linear_regression(prevaccine_df)

print("\nPost-vaccine period:")
perform_linear_regression(postvaccine_df)
```

**Pre-vaccine period:**

The coefficients for depression and anxiety are -7.78318695 and 4.04060493, respectively, and the intercept is about 3423.74 grams.
This suggests that, on average, an increase of one unit in depression scores is associated with a decrease of approximately 7.78 grams in birth weight. Similarly, an increase of one unit in anxiety scores is associated with an increase of approximately 4.04 grams in birth weight during the pre-vaccine period.

**Post-vaccine period:**

The coefficients for depression and anxiety are -3.23921234 and -0.86587632, respectively, with an intercept of about 3443.02 grams.
This implies that, on average, an increase of one unit in depression scores is associated with a decrease of approximately 3.24 grams in birth weight. Conversely, an increase of one unit in anxiety scores is associated with a decrease of approximately 0.87 grams in birth weight during the post-vaccine period.

The limitations of this study are important for understanding the context of our findings and guiding future research. One significant limitation is the disparity in the number of observations between the pre-vaccine (2,923) and post-vaccine (65) subsets. This substantial difference could potentially skew the analysis, making it challenging to draw definitive conclusions about the impact of the vaccine rollout on pregnant women's mental health and childbirth outcomes. The smaller sample size in the post-vaccine period limits the statistical power of our analysis, making it more difficult to detect significant differences or trends that may exist.
	
Another limitation is the study's reliance on self-reported data, which may introduce bias due to participants' subjective interpretation of their mental health status and experiences during the pandemic. Additionally, the study's focus on a Canadian population may limit the generalizability of the findings to other regions with different healthcare systems, pandemic responses, and vaccination rollouts. When we looked at our population data during our EDA, we found that our dataset lacked diversity in educational background, ethnicity, language spoken, and socioeconomic status, as the majority of our participants were white, from middle-class income homes, who held degrees and spoke English. This gives the applicability of our findings to be limited to the scope of the Canadian population which it was sampled from, lacking the ability to be compared to other countries and demographics of people. Though our group tried hard to find widespread published maternity mental health data over the pandemic, none had the multitude of variables and components of established research built into it like this dataset, which is why we made it our primary dataset. 
	
The timing and context of the vaccine rollout also represent a limitation. The initial phase of vaccine distribution prioritized certain populations, and pregnant women's eligibility and willingness to receive the vaccine during our post-vaccine study period may have varied widely. This variability could influence the observed outcomes and their association with mental health indicators.
	
Future research should aim to address these limitations by including larger and more diverse post-vaccine samples to enhance the generalizability of the findings. Longitudinal studies that follow pregnant women through the pre-vaccine, vaccine rollout, and post-vaccine phases could provide more comprehensive insights into the pandemic's long-term effects on maternal mental health and childbirth outcomes. Additionally, investigating other factors that may influence these outcomes, such as healthcare access, socioeconomic status, and vaccination status, would offer a more nuanced understanding of the pandemic's impact on pregnant women. Further research could also explore interventions designed to mitigate the adverse effects of pandemic-related stress on pregnant women and their babies, thereby informing public health strategies and healthcare practices to better support this vulnerable population during global health crises.
